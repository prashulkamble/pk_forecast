## 1. processing of prices data

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pdcast as pdc
import os
for dirname, _, filenames in os.walk(dirct):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
prices = pd.read_csv(dirct + 'sell_prices.csv')
prices = pdc.downcast(prices)
```

```python
prices = prices.sort_values(by=['item_id','store_id','wm_yr_wk'])
```

```python
prices.sample(10)
```

```python
path2 = r'D:\python_code\M5 Competition\A5_solution\A5_output2\\'
for win in [1, 2, 4, 8]:
    prices['rm_diff_price_{}'.format(win)] = prices[["store_id","item_id","sell_price"]].groupby(
        ["store_id","item_id"])["sell_price"].transform(lambda x : x.rolling(win).mean())
    prices['rm_diff_price_{}{}'.format(win,win)] = ((prices['sell_price'] - prices['rm_diff_price_{}'.format(win)])/prices['sell_price']).round(3)
prices.to_csv(path2 + 'prices_all.csv')
```

## 2. Processing of calender data

```python
import numpy as np 
import pandas as pd
import pdcast as pdc

import os
for dirname, _, filenames in os.walk(dirct):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
calendar = pd.read_csv(dirct + 'calendar.csv')
calendar = pdc.downcast(calendar)
```

```python
calendar.head()
```

```python
dict_day_to_group = {
    1:1, 2:1, 3:1, 4:1, 5:1,
    6:2, 7:2, 8:2, 9:2, 10:2,
    11:3, 12:3, 13:3, 14:3, 15:3,
    16:4, 17:4, 18:4, 19:4, 20:4,
    21:5, 22:5, 23:5, 24:5, 25:5,
    26:6, 27:6, 28:6, 29:6, 30:6, 31:6  
}

calendar['group_day'] = calendar['date'].str[-2:].apply(int).map(dict_day_to_group)
# calendar = calendar.drop(['date','weekday'], axis=1)
calendar = calendar.rename(columns={'d':'day'})
```

```python
calendar = pd.concat([calendar, pd.get_dummies(calendar[['event_name_1','event_name_2']])], axis=1)
calendar = calendar.drop(['event_name_1','event_name_2','event_type_1','event_type_2'], axis=1)
```

```python
calendar
```

```python
for col in [c for c in calendar.columns.tolist() if 'event_name' in c]:
    days_event = np.where(calendar[col] == 1)[0].tolist()
    print(col, np.where(calendar[col] == 1))
    col2 = col + '_analysis'
    calendar[col2] = calendar['day']
    col3 = col + '_analysis2'
    dict_days_event = {}
    for d in days_event:
        for i in range(0, 30):
            dict_days_event['d_'+str(d-i)] = 30-i
    calendar[col3] = calendar[col2].map(dict_days_event).fillna(0)
```

```python
for col in [c for c in calendar.columns.tolist() if 'event_name' in c]:
    days_event = np.where(calendar[col] == 1)[0].tolist()
    calendar[col] = calendar['day']

    dict_days_event = {}
    for d in days_event:
        for i in range(0, 30):
            dict_days_event['d_'+str(d-i)] = 30-i

    calendar[col] = calendar[col].map(dict_days_event).fillna(0)
```

```python
calendar['event_sum'] = calendar[[c for c in calendar.columns.tolist() if 'event_name' in c]].sum(axis=1).tolist()
```

```python
calendar.replace({True: 1, False: 0}, inplace= True)
```
