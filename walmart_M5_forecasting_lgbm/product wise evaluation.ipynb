```python
dirct = r'D:\python_code\M5 Competition\m5-forecasting-Dataset\\'
path2 = r'D:\python_code\M5 Competition\A5_solution\A5_output2\\'
import pandas as pd
```

```python
merge_data  = pd.read_csv(path2 + 'merged_data.csv')
factors = pd.read_csv(r'D:\python_code\M5 Competition\A5_solution\A5_output\factor_18nov.csv')
```

```python
del merge_data
```

```python
combine_data = merge_data.merge(factors, on = ['dept_id', 'store_id'], how = 'left')
combine_data = combine_data.drop_duplicates()
combine_data = combine_data.reset_index(drop=True)
combine_data
```

```python
# apply factors
cols_f = ['F{}'.format(c) for c in range(1, 29)]
for c in cols_f:
    combine_data[c] = combine_data[c]*combine_data['factor']
combine_data
```

```python
cols_f = ['F{}'.format(c) for c in range(1, 29)]
cols_d = [f'd_{i}' for i in range(1914, 1942)]
cols_a = [f'd_{i}_A' for i in range(1914, 1942)]
cols_p = [f'd_{i}_P' for i in range(1914, 1942)]

processed_data = combine_data.rename(columns=dict(zip(cols_d,cols_a))).rename(columns=dict(zip(cols_f,cols_p)))
processed_data["id"] = processed_data["id"].str.replace("_validation","")
processed_data.drop('factor', axis= 1, inplace=True)



```

```python
processed_data['actual_sale_28day_sum'] = processed_data[cols_a].sum(axis=1)
processed_data['predicted_sale_28day_sum'] = processed_data[cols_p].sum(axis=1)
```

```python
processed_data['actual_sale_28day_avg'] = processed_data[cols_a].mean(axis=1)
processed_data['predicted_sale_28day_avg'] = processed_data[cols_p].mean(axis=1)
```

```python
processed_data['diff_sale_28day_sum'] = processed_data['predicted_sale_28day_sum'] - processed_data['actual_sale_28day_sum']
processed_data['diff_sale_28day_avg'] = processed_data['predicted_sale_28day_avg'] - processed_data['actual_sale_28day_avg']
```

```python
processed_data['positive_pred'] = processed_data['diff_sale_28day_sum'].apply(lambda x : '+' if x >0 else '-')
```

```python
# MAPE
import numpy as np
processed_data['APE'] = (np.abs((processed_data['actual_sale_28day_sum'] - processed_data['predicted_sale_28day_sum']) / processed_data['actual_sale_28day_sum']))
processed_data['APE_%'] = (np.abs((processed_data['actual_sale_28day_sum'] - processed_data['predicted_sale_28day_sum']) / processed_data['actual_sale_28day_sum']))* 100
```

```python
#SMAPE
processed_data['SMAPE'] = (np.abs((processed_data['actual_sale_28day_sum'] - processed_data['predicted_sale_28day_sum']) /(((processed_data['actual_sale_28day_sum'] + processed_data['predicted_sale_28day_sum'])/2))))
```

```python
processed_data.to_csv(path2 + 'processed_data_merged.csv')
```
