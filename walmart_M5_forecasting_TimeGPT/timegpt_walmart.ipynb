## Forecasting M5 data with timegpt

```python
import os
import pandas as pd
import numpy as np
import pdcast as pdc
import matplotlib.pyplot as plt
import gc
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_absolute_error, mean_squared_error
import joblib
```

```python
sales = pd.read_csv(dirct + 'sales_train_evaluation.csv')                             ## added
sales.name = 'sales'
calendar = pd.read_csv(dirct +'calendar.csv')
calendar.name = 'calendar'
prices = pd.read_csv(dirct +'sell_prices.csv')
prices.name = 'prices'
```

```python
sales = pdc.downcast(sales)
calendar = pdc.downcast(calendar)
prices = pdc.downcast(prices)
```

## Forecast with finetuning

```python
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display, clear_output
forecast_df_tune = pd.DataFrame()
cnt = 0
for i in item_unique:
    clear_output(wait=True)
    display(f'forecast for={i} and count = {cnt}/{len(item_unique)}')
    product_id = [prod for prod in products if i in prod]
    Y_df_product = Y_df.query('unique_id in @product_id').reset_index(drop=True)
    X_df_product = X_df.query('unique_id in @product_id').reset_index(drop=True)
    timegpt_fcst_df = nixtla_client.forecast(df=X_df_product, X_df=Y_df_product, h=28, finetune_steps=20, finetune_loss='mse')
    timegpt_fcst_df['ds'] = timegpt_fcst_df['ds'].astype({'ds': 'datetime64[ns]'})
    del Y_df_product, X_df_product
    forecast_df_tune = pd.concat([forecast_df_tune, timegpt_fcst_df], axis= 0)
    del timegpt_fcst_df
    cnt += 1
```

```python
forecast_df_tune.to_csv('/teamspace/studios/this_studio/M5_forecast_transform/ '+ 'forecast_df_tune.csv')
```
